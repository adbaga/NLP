<h3>1. Function Synsets

```python
#Print all synonyme of Love
syn = wn.synsets('Love')
print(syn)
```

Synset() function presents in NLTK to look up words in WordNet. It will return a list of all the possible synonym of the word that we request. It will give us what class of word it is, for example 'Love' can be a noun and even the noun of love can have several definition. It definite;y has different meaning than love as a verb

```python
#Print Definitin, Lemma, and Examples of the first synonym

print('Definition:',syn[0].definition())
print('Lemma:',syn[0].lemmas())
print('Examples:',syn[0].examples())
```

Definition: This function will give the meaning / definition of this particular synonym<br>

Lemma: This function is used as a result of lemmatizartion proces lemmatization will remove inflectional endings only and to return the base or dictionary form of a word  which is known as the lemma<br>

Examples: Examples will give an example of how this synonym of word is used in a sentence


```python
#Print synonyms and antonyms of the word 'Love'

syn = []
ant = []

word = "Love"
for synset in wn.synsets(word):
   for lemma in synset.lemmas():
      syn.append(lemma.name())
      if lemma.antonyms():
          ant.append(lemma.antonyms()[0].name())
print(f'Synonyms of {word}: ' + str(syn), '\n')

print(f'Antonyms of {word}: ' + str(ant), '\n')
```

<h3>4. Compute the similarity score between pairs of words

```python
cat = wn.synset('cat.n.01')
dog = wn.synset('dog.n.01')
car = wn.synset('car.n.01')

catDogPS = cat.path_similarity(dog)
catDogLS = cat.lch_similarity(dog)
catDogWS = cat.wup_similarity(dog)


dogCarPS = dog.path_similarity(car)
dogCarLS = dog.lch_similarity(car)
dogCarWS = dog.wup_similarity(car)

catCarLS = cat.lch_similarity(car)
catCarPS = cat.path_similarity(car)
catCarWS = cat.wup_similarity(car)

print(f'Cat-Dog Path Similarity: {catDogPS} --- LCH Similarity: {catDogLS} --- WUP Similarity: {catDogWS }\n')

print(f'Dog-Car Path Similarity: {dogCarPS} --- LCH Similarity: {dogCarLS} --- WUP Similarity: {dogCarWS }\n')

print(f'Cat-Car Path Similarity: {catCarPS} --- LCH Similarity: {catCarLS} --- WUP Similarity: {catCarWS }\n')



```

Path Similarity: Return a score of how similar two words are based on the shortest path that connects them in their word taxonomy

Leacock-Chodorow Similarity: Giving a score of two words similarity, based on the shortest path that connects the senses and the maximum depth of the taxonomy of the senses. It can be defined as -log(p/2d) (p is the shortest path length and d the depth of the taxonomy)

Wu-Palmer Similarity: Giving score of two words similarity, it is based on the depth of the two senses in the taxonomy and that of their Least Common Subsumer (most specific ancestor node).


From the Path similarity, Leacock-Chodorow similarity, and Wu-Palmer Similarity scores we can conclude that Cat and Dog has more similarity than the other two pairs. Both are more similar by definition (since they both are animal) and this measure is not made based on how the word is similar synthatically

<h1>D. TF-IDF Analysis with TextBlob and Wikipedia

```python
corpus = []
topics = ['France', 'Elton John', 'Python', 'Fox', 'Isaac Newton', 'Zinedine Zidane']

for topic in topics:
    corpus.append(TextBlob(wikipedia.summary(topic)))
```

```python
def tf(word, top):
    freq = 0
    text = top.words
    noWords = len(text)
    for i in range(0,noWords):
        if word == text[i]:
            freq +=1
    #print('The word',word, 'shows up', freq,'times')
    #print(freq)
    return freq/noWords

def n_containing(word, listTop):
    freq = 0
    lenLT = len(listTop)
    wordText = []
    for i in range(0,lenLT):
        wordText.append(listTop[i].words)

    for j in range(0,len(wordText)):
        if word in wordText[j]:
            freq += 1

    #print('The word',word, 'appears in', freq, 'topics')
    return freq

def idf(word, listTop):
    x = len(listTop)
    appearsIn = n_containing(word, listTop)

    idf = np.log(x/(1+appearsIn))
    #print(idf)
    
    return idf

def tfidf(word, top, listTop):
    tfVal = tf(word,top)
    idfVal = idf(word, listTop)
    tfidfVal = tfVal*idfVal
    return tfidfVal
```

TF (or term frequency) has many different formulas. In this lab, we count the frequency of the word in a text and how often that word is mentioned compared to how big the document is in terms of number of word.

IDF measures the importance of document in whole set of corpus, but instead of the frequency, it measures the existance of a word inside documents in the text regardless of how many time that term is mentioned in one document. It will stay be counted as one. In my IDF formula, I added 1 in the number of document which the term appears to avoid division by zero


The high TF score indicate how one word has a significant role in the text. Then, we use IDF it normalize it and compare it with the other topics in the corpus. TF-IDF is merely just the value of TF multiplied by the value IDF

```python
#Cleaning from stop words and make all lower case
stop_words=set(stopwords.words("english"))
freq = 0
lenLT = len(corpus)
cleanCorpus = []
for i in range(0,lenLT):
    if i not in stop_words:
        cleanCorpus.append((corpus[i]).lower())
```

```python
#defining document name

france = cleanCorpus[0]
eltonJohn = cleanCorpus[1]
python = cleanCorpus[2]
fox = cleanCorpus[3]
isaacNewton = cleanCorpus[4]
zinedineZidane = cleanCorpus[5]
```

```python
def maxTFIDF(doc):
    tfidfScr = []
    for i in doc.words: 
        tfidfScr.append(tfidf(i,doc,cleanCorpus))
    maxScr = max(tfidfScr)
    maxScrWord = doc.words[tfidfScr.index(maxScr)]
    print(maxScr)
    print(maxScrWord)
    return maxScr, maxScrWord

```

```python
franceMax = maxTFIDF(france)
eltonJohnMax = maxTFIDF(eltonJohn)
pythonMax = maxTFIDF(python)
foxMax = maxTFIDF(fox)
isaacNewtonMax = maxTFIDF(isaacNewton)
zinedineZidaneMax = maxTFIDF(zinedineZidane)
```
